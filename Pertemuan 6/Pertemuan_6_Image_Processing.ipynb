# **Perhitungan Matrix 3x3**

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# 1. Input matriks citra
img = np.array([
    [148, 164, 174],
    [144, 159, 167],
    [128, 147, 157]
], dtype=float)

print("Citra Asli:")
print(img)
```

```python
# 2. Hitung FFT (DFT 2D)
f = np.fft.fft2(img)

print("\nHasil DFT:")
print(f)
```

```python
# 3. Shift frekuensi ke tengah
fshift = np.fft.fftshift(f)

print("\nDFT setelah di-shift:")
print(fshift)
```

```python
# 4. Hitung magnitude spectrum
magnitude = np.abs(fshift)

print("\nMagnitude Spectrum:")
print(magnitude)
```

```python
# 5. Visualisasi
plt.figure(figsize=(10,4))

# Citra asli
plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Citra Asli')
plt.colorbar()

# Magnitude spectrum
plt.subplot(1,2,2)
plt.imshow(magnitude, cmap='gray')
plt.title('Magnitude Spectrum')
plt.colorbar()

plt.tight_layout()
plt.show()
```
