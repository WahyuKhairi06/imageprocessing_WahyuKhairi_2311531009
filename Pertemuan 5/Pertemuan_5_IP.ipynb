```python
import numpy as np
import matplotlib.pyplot as plt


# 1. Load Citra

img = Image.open('/content/dog.jpg').convert('L') # Convert ke Grayscale
img_array = np.array(img)

# 2. Definisi Kernel Rataan 5x5 (Mean Filter)
# Sesuai slide: nilai tiap elemen adalah 1/25
kernel_size = 5
kernel = np.ones((kernel_size, kernel_size)) / (kernel_size**2)

def manual_convolution(image, kernel):
    img_h, img_w = image.shape
    ker_h, ker_w = kernel.shape

    # Menentukan ukuran output (tanpa padding)
    output_h = img_h - ker_h + 1
    output_w = img_w - ker_w + 1
    output = np.zeros((output_h, output_w))

    # Proses Konvolusi Manual
    for i in range(output_h):
        for j in range(output_w):
            # Ambil area sub-matriks sesuai ukuran kernel
            region = image[i:i+ker_h, j:j+ker_w]
            # Operasi perkalian elemen dan penjumlahan
            output[i, j] = np.sum(region * kernel)

    return output

# 3. Eksekusi Fungsi
output_img = manual_convolution(img_array, kernel)

# 4. Visualisasi Hasil
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.title("Citra Asli (Grayscale)")
plt.imshow(img_array, cmap='gray')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.title("Hasil Konvolusi (Kernel Rataan 5x5)")
plt.imshow(output_img, cmap='gray')
plt.axis('off')

plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 1. Load Citra Baru

# Load RGB image
img_baru_rgb = Image.open('/content/profile.jpg') # Load in RGB format
print('/content/profile.jpg')

# Convert to Grayscale for convolution
img_baru= img_baru_rgb.convert('L')
img_array_baru = np.array(img_baru)

# 2. Tetap menggunakan Kernel Rataan 5x5 sesuai instruksi tugas
kernel_size = 5
kernel_mean = np.ones((kernel_size, kernel_size)) / (kernel_size**2)

# 3. Eksekusi Konvolusi Manual (Memanggil fungsi yang sudah dibuat sebelumnya)
output_baru = manual_convolution(img_array_baru, kernel_mean)

# 4. Normalisasi: Memastikan nilai piksel dalam range yang benar untuk visualisasi
output_normalized = np.clip(output_baru, 0, 255).astype(np.uint8)

# 5. Visualisasi Hasil Baru
plt.figure(figsize=(21, 7)) # Increased figure size for 3 plots

plt.subplot(1, 3, 1) # First subplot for RGB image
plt.title("Citra Baru (Original RGB)")
plt.imshow(img_baru_rgb)
plt.axis('off')

plt.subplot(1, 3, 2) # Second subplot for Original Grayscale
plt.title("Citra Baru (Original Grayscale)")
plt.imshow(img_array_baru, cmap='gray')
plt.axis('off')

plt.subplot(1, 3, 3) # Third subplot for Convoluted Image
plt.title("Hasil Smoothing 5x5 (Citra Baru)")
plt.imshow(output_normalized, cmap='gray')
plt.axis('off')

plt.tight_layout()
plt.show()

# Jika ingin mendownload hasil untuk laporan:
# Image.fromarray(output_normalized).save('hasil_konvolusi_5x5.jpg')
```
